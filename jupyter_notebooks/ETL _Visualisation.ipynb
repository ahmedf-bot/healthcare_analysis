# Section 2

Section 2 content

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['charges'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Insurance Charges ($)')
plt.title('Impact of Age on Medical Insurance Cost')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.boxplot(column='charges', by='region')
plt.ylabel('Insurance Charges ($)')
plt.title('Impact of Region on Medical Insurance Cost')
plt.suptitle('')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
smoker_yes = df[df['smoker'] == 'yes']['charges']
smoker_no = df[df['smoker'] == 'no']['charges']

plt.boxplot([smoker_no, smoker_yes], labels=['Non-Smoker', 'Smoker'])
plt.ylabel('Insurance Charges ($)')
plt.title('Impact of Smoking on Medical Insurance Cost')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['bmi'], df['charges'], alpha=0.5)
plt.xlabel('BMI')
plt.ylabel('Insurance Charges ($)')
plt.title('Impact of BMI on Medical Insurance Cost')
plt.grid(True, alpha=0.3)
plt.show()
```

---

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['age'], df['bmi'], c=df['charges'], cmap='viridis', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Correlation between Age, BMI, and Medical Insurance Cost')
cbar = plt.colorbar(scatter)
cbar.set_label('Insurance Charges ($)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='charges', hue='smoker', data=df)
plt.title('Impact of Sex and Smoking Status on Medical Insurance Cost')
plt.ylabel('Insurance Charges ($)')
plt.xlabel('Sex')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='children', y='charges', data=df)
plt.xlabel('Number of Children')
plt.ylabel('Insurance Charges ($)')
plt.title('Impact of Number of Children on Medical Insurance Cost')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df, alpha=0.7)
plt.xlabel('BMI')
plt.ylabel('Insurance Charges ($)')
plt.title('Correlation of BMI and Smoking Status on Medical Insurance Cost')
plt.grid(True, alpha=0.3)
plt.legend(title='Smoker')
plt.show()
```
